# ENTREGA WEB SACRAP

El objetivo de esta entrega es realizar webscraping de una busqueda en mercadolibre sobre un determinado modelo de auto para hacer una clasificacion segun los precios, modelo, a√±o, km. Luego de este analisis podremos determinar cual es el precio acorde para publicar un auto de similares caracteristicas.

```python
url = requests.get('https://listado.mercadolibre.com.ar/peugeot-208-feline-2018')
url.status_code
```

```python
#check (No hace falta hacerlo)
url.content
```

```python
soup = BeautifulSoup(url.content, 'html.parser')
```

```python
titulos = soup.find_all('h2', attrs={"class":'ui-search-item__title ui-search-item__group__element shops__items-group-details shops__item-title'})
```

```python
titulos[0].text
```

```python
#(No hace falta hacerlo)
for i in titulos:
    
    print(i.text)
```

```python
titulos = [i.text for i in titulos]
```

```python
titulos
```

```python
len(titulos)
```

```python
urls = soup.find_all('a', class_="nav-header nav-header-plus ui-navigation-v2")
```

```python
urls[0].get('href')
```

```python
urls = [i.get('href') for i in urls]
urls
```

```python
print(len(titulos))
len(urls)
```

```python
df = pd.DataFrame({'titulos':titulos, 'url':urls})
df['fecha_obs'] = date.today()
df
```

```python
precio = soup.find_all('div', class_="ui-search-item__group__element ui-search-price__part-without-link")

precio
```

```python
precio = [p.text for p in precio]
```

```python
#precio = [p[12:] for p in precio]
precio
```

```python
len(precio)
```

```python
df = pd.DataFrame({'titulos':titulos, 'url':urls, 'precio':precio })
df['fecha_obs'] = date.today()
df
```
