```python
import time
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.dates import DateFormatter
   
# Lee el archivo CSV en un DataFrame
df_usdt = pd.read_csv('datos_usdt.csv')


df_btcusdt = df_usdt[df_usdt['Criptomoneda'] == 'BTCUSDT']
df_ethusdt = df_usdt[df_usdt['Criptomoneda'] == 'ETHUSDT']
df_trxusdt = df_usdt[df_usdt['Criptomoneda'] == 'TRXUSDT']

# Muestra los primeros 100 registros de cada DataFrame por separado
df_btcusdt.head(200)

```

```python
df_trxusdt.head(100)
```

```python
df_ethusdt.head(100)
```

```python
# Configurar el gráfico para BTCUSDT
plt.figure(figsize=(10, 6))
plt.plot(df_btcusdt['Minutos'], df_btcusdt['Precio'], label='BTCUSDT', color='blue')


plt.title('Precio de BTCUSDT')
plt.xlabel('Minutos')
plt.ylabel('Precio')
plt.xticks(rotation=45)
plt.legend()
plt.tight_layout()
plt.show()
```

```python

# Configurar el gráfico para ETHUSDT
plt.figure(figsize=(10, 6))
plt.plot(df_ethusdt['Minutos'], df_ethusdt['Precio'], label='ETHUSDT', color='green')


plt.title('Precio de ETHUSDT')
plt.xlabel('Minutos')
plt.ylabel('Precio')
plt.xticks(rotation=45)
plt.legend()
plt.tight_layout()
plt.show()


```

```python
# Graficar los precios de TRXUSDT
plt.plot(df_trxusdt['Minutos'], df_trxusdt['Precio'], label='TRXUSDT')
plt.title('Precio de TRXUSDT')
plt.xlabel('Minutos')
plt.ylabel('Precio')
plt.xticks(rotation=45)
plt.legend()
plt.show()
```
