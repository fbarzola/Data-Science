```python
import requests
import pandas as pd
import numpy as np
import seaborn
from matplotlib import pyplot as plt
import matplotlib 
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import warnings
warnings.filterwarnings("ignore")
import spotipy
from spotipy.oauth2 import SpotifyClientCredentials
```

```python
#devoloper.spotify.com

CLIENT_ID = ''
CLIENT_SECRET = ''
```

```python
client_credentials_manager = SpotifyClientCredentials(client_id=CLIENT_ID, client_secret=CLIENT_SECRET)
sp = spotipy.Spotify(client_credentials_manager=client_credentials_manager)
```

```python
resultado = sp.search(q='artist:The Beatles', type='artist')
resultado
```

```python
artist_id = resultado['artists']['items'][0]['id']

artist_id
```

```python
artist_info = sp.artist(artist_id)
artist_info
```

```python
print('Nombre: ', artist_info['name'])
print('Imagen: ', artist_info['images'][0]['url'] )
print('Seguidores: ', artist_info['followers']['total'] )
print('Generos: ', artist_info['genres'])
```

```python
top = sp.artist_top_tracks(artist_id)
top['tracks'][0]['name']
```

```python
for i in range(0,10):
    for t in top:
        print(top['tracks'][i]['name'])
        print(round(top['tracks'][i]['duration_ms']/60000,2))
        print(top['tracks'][i]['track_number'])
```

```python
#armar un df album, fecha de lanz, tracks, link a la portada

albums = sp.artist_albums(artist_id, album_type = 'album',limit=50)

album = []
lanzamiento = []
tracks = []
portada = []


for a in albums['items']:
    album.append(a['name'])
    lanzamiento.append(a['release_date'])
    tracks.append(a['total_tracks'])
    portada.append(a['images'][0]['url'])
    
len(album),len(lanzamiento),len(tracks),len(portada) 
```

```python
df = pd.DataFrame({'album': album, 'lanzamiento': lanzamiento, 'tracks': tracks,'portada': portada })
df
```
