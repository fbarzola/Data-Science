```python
import pandas as pd
import matplotlib.pyplot as plt

data = [
    ["Ana Rodriguez", "Elena Castro", "Dólares", 177.60, 1, 177.60],
    ["Maria Gonzalez", "Ricardo Torres", "Dólares", 285.14, 10, 2851.40],
    ["Carlos Lopez", "Elena Castro", "Dólares", 454.45, 8, 3635.60],
    ["Carlos Lopez", "Ricardo Torres", "Dólares", 268.35, 8, 2146.80],
    ["Juan Pérez", "Pedro Sanchez", "Dólares", 149.59, 5, 747.95],
    ["Maria Gonzalez", "Elena Castro", "Pesos", 335.97, 10, 3359.70],
    ["Ana Rodriguez", "Elena Castro", "Dólares", 282.55, 6, 1695.30],
    ["Maria Gonzalez", "Sofia Ramirez", "Dólares", 241.32, 6, 1447.92],
    ["Maria Gonzalez", "Ricardo Torres", "Pesos", 495.97, 1, 495.97],
    ["Luis Fernandez", "Pedro Sanchez", "Dólares", 432.38, 3, 1297.14]
]

df = pd.DataFrame(data, columns=["Cliente", "Vendedor", "Moneda", "PrecioUnitario", "Cantidad", "Total"])

print("Datos de ventas:")
print(df)

ganancias_por_vendedor = df.groupby("Vendedor")["Total"].sum()

plt.figure(figsize=(8, 8))
plt.pie(ganancias_por_vendedor, labels=ganancias_por_vendedor.index, autopct='%1.1f%%', startangle=140)
plt.title("Distribución de Ganancias por Vendedor")
plt.show()
```
